```python
def all_songs(album_ids):
    song_list=[]
    song_id_list=[]
    for i in album_ids:
        bresults =sp.album_tracks(i, limit=50, offset=0, market=None)
        for j in bresults['items']:
            song_list.append(j['name'])
            song_id_list.append(j['id'])
    return song_list, song_id_list
```
