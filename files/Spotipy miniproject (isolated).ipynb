```python
def get_artist_id(artist_list):
    artist_id_list=[]
    missing_ids=[]
    for i in artist_list:
        results = sp.search(q=i, limit=50)
        l=0
        c=0
        for j in results['tracks']['items']:
            l+=1
            results = sp.search(q=i, limit=50)
            if (j['artists'][0]['name'].lower() == i.lower())&(c==0):
                artist_id_list.append(results['tracks']['items'][0]['artists'][0]['id'])
                c+=1
            if (c==0)&(l==49):
                missing_ids.append(i)
    return artist_id_list, missing_ids
```
